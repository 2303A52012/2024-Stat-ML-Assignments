## **reading data**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
sal = pd.read_csv('Salary_Data.csv')
sal
```

## **features and target variables**

### **target variable**

```python
y=sal['Salary']
print(f"the target variable is {y.name} \nand the data is \n")
print(y)
print(f"the range of target variable is {y.min()} - {y.max()}")
```

### **features**

```python
x=sal['YearsExperience']
print(f"the feature is {x.name} \nand the data is \n")
print(x)
print(f"the range of target variable is {x.min()} - {x.max()}")
```

## **graph**

```python
sal.plot(kind='scatter', x='YearsExperience', y='Salary', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
plt.plot(x,y,color='blue',label='Salary',linewidth=2)
plt.title("experience V/S salary",fontsize=16,
          fontstyle='oblique')
plt.xlabel("Years of Experience",fontsize=15)
plt.xticks(fontsize=13,rotation=0)
plt.yticks(fontsize=13)
plt.ylabel("Salary",fontsize=15)
plt.show()
```

```python

plt.title("experience V/S salary",fontsize=16,
          fontstyle='oblique')
plt.bar(x=x,height=y, width=0.5,
        color='blue',edgecolor='black',alpha=1)
plt.xlabel("Years of experiences",fontsize=15)
plt.xticks(fontsize=13,rotation=35,color="red")
plt.yticks(fontsize=13)
plt.ylabel("Salary",fontsize=15)
plt.show()
```

## **Normalize the give data[0 to 1]**

```python
Norm_data = (sal-sal.min())/(sal.max()-sal.min())
Norm_data
```
